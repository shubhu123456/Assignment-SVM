```python
from sklearn.svm import SVC
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
data=pd.read_csv("C:\\Excelr files\\forestfires.csv")
```

```python
data
```

```python
col=list(data.columns)
```

```python
col
```

```python
data.drop(['month','day'],axis=1,inplace=True)
```

```python
data
```

```python
data.tail()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
data['size_category']=label_encoder.fit_transform(data['size_category'])
```

```python
data
```

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
data_norm=norm_fun(data)
```

```python
data_norm
```

```python
col_norm=list(data_norm.columns)
```

```python
col_norm
```

```python
array=data_norm.values
```

```python
X=array[:,0:28]
y=array[:,28]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=40)
```

```python
cif=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
grid=GridSearchCV(cif,param_grid,cv=10)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
cif=SVC(gamma=50,C=0.1)
```

```python
cif.fit(X_train,y_train)
```

```python
y_pred=cif.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_pred,y_test)
```

```python
accuracy*100
```

```python
#accuracy is 75.96%
```
